axis=column when you feed it with(1) when you feed it with (0) then it refers to rows

0.5 refers to 50% any thing above 50% will be dropped  since that means many fveatures are missing

```python
data.dropna(axis=1,thresh=0.5*len(data),inplace=True)
```

```python
data['MasVnrType']
```

```python
mean_fill=['LotFrontage','MasVnrArea']
bfill=['FireplaceQu']
ffill=(set(features_with_missing_values)-set(mean_fill)-set(bfill))
```

```python
ffill
```

```python
for feature in mean_fill:
  data[feature].fillna(data[feature].mean(),inplace=True)
```

```python
for feature in ffill:
  data[feature].fillna(method='ffill',inplace=True)
```

```python
for feature in ffill:
  data[feature].fillna(method='bfill',inplace=True)
```

```python
data.isna().any()
```

```python
#pca1 carries 10% of information about data while pca2 carries 4% of information about dataset
pca.explained_variance_ratio_
```

```python
#this is another way of ensuring pca capture 90% of information about dataset
pca_var=PCA(0.9)
new_pcas=pca_var.fit_transform(X)
```

```python
pca_data=pd.DataFrame(p_components,columns=['PC1','PC2'])
target_data=pd.DataFrame(y,columns=['SalePrice'])
final_data=pd.concat([pca_data,target_data],axis=1)
final_data
```

# Model Selection

```python

result={}
models={
   'LinearRegression': linear_model.LinearRegression(),
    'Laso':linear_model.Lasso(),
    'Ridge':linear_model.Ridge(),
    'ElasticNet':linear_model.ElasticNet()
}
#modelli=linear_model.LinearRegression()
#modelli.fit(X_train,y_train)
#modelli.fit(X,y)
#accuracy=modelli.score(X_test,y_test)

for i in models:
  models[i].fit(X_train,y_train)
  accuracy=models[i].score(X_test,y_test)
  result[i]=[accuracy]
  pd.DataFrame(result)
```
